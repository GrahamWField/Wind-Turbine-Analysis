# Creating the Correlation Matrix Heatmap

```python
# Calculate the correlation matrix
corr_matrix = df_corr.corr()
```

```python
# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Looking at the Correlations

##### Strong Correlations
Gross performance, Net rated power, Hub Height, and Rotor Diameter have very strong correlations which makes sense, as the turbine technology improves the turbines have been getting taller, their blades getting longer, and their power outputs have been increasing.
with relation to Rotor Diameter, we can see this very well.
it is worth noting here that net rated power and gross output are the same and that is why we have a 1-1 correlation.

with hub height and power, we see a very strong correlation, this makes sense as the larger the blades are the more wind they can catch and the more power they can generate.

with hub height and rotor diameter, we also see a strong correlation which also makes sense as the longer the blades get the higher you need to build the turbine. we don't want blades hitting the ground and we also want to keep the blades in the wind.

we can also see a correlation between water depth and distance from the coastline which makes sense, as the further you go from land the deeper the water gets.

##### Weak Correlations
when it comes to the offshore sector. we see weak correlations because this does not determine the type of turbine that is built. As the technology improves the biggest and the best will be built.
given that some turbines can be built to float the water depth and distance from the coastline really should have much impact. water depth and distance from the coastline also have weak correlations and would have very little to do with the type of turbine built as the water line is the water line and the metrics we have are ultimately determined by that. 
we can say that the the deeper the water and the further away from the coastline does not correlate to larger more powerful turbines.

##### Conclusion

the improvement of technology is represented by the growth in power output and the strong correlating increase in Rotor diameter and hub height. As the turbines become more powerful the blades get longer and the towers get taller. the turbines being built offshore does not impact the improved power and size of the turbines and we can see this in the weak correlation. with water depth and distance from shore.

```python
# Convert 'Commissioning Date' to datetime format
df_gwt['Commissioning Date'] = pd.to_datetime(df_gwt['Commissioning Date'])

# Extract the year component from the 'Commissioning Date' column
df_gwt['Year'] = df_gwt['Commissioning Date'].dt.year

# Create a pivot table to aggregate counts of occurrences of each combination of State and Year
heatmap_data = df_gwt.groupby(['State', 'Year']).size().unstack(fill_value=0)

# Create the heatmap with grid
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', cbar_kws={'label': 'Frequency'}, linewidths=0.5)
plt.title('Heatmap of State and Commissioning Year')
plt.xlabel('Commissioning Year')
plt.ylabel('State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Calculate the sum of Gross Performance for each manufacturer
manufacturer_gross_performance_sum = df_gwt.groupby('Manufacturer')['Gross Performance'].sum()

# Select the top 10 manufacturers based on Gross Performance sum
top_10_manufacturers = manufacturer_gross_performance_sum.nlargest(10)

# Create a colormap with a variety of blues
cmap = plt.get_cmap('Blues')

# Plot the results with different shades of blue
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_manufacturers.index, top_10_manufacturers.values, color=cmap(np.linspace(0.2, 0.8, len(top_10_manufacturers))))
plt.title('Sum of Gross Performance by Top 10 Manufacturers')
plt.xlabel('Manufacturer')
plt.ylabel('Sum of Gross Performance')
plt.xticks(rotation=45, ha='right')

# Set y-axis tick labels format to display full numbers
plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

plt.tight_layout()
plt.show()
```

### Summary

Here we can see that Enercon has the largest production of all manufacturers followed by Vestas and then Siemens given that bigger turbines mean more powerful turbines it would be interesting to see how the manufacturers compare when looking at their ranges of total power outputs.

```python
# Get the top 10 manufacturers by gross performance
top_10_manufacturers = df_gwt.groupby('Manufacturer')['Gross Performance'].sum().nlargest(10).index

# Filter the DataFrame to include only the top 10 manufacturers
df_top_10 = df_gwt[df_gwt['Manufacturer'].isin(top_10_manufacturers)]

# Convert 'Commissioning Date' to datetime format
df_top_10['Commissioning Date'] = pd.to_datetime(df_top_10['Commissioning Date'])

# Extract the year component from the 'Commissioning Date' column
df_top_10['Year'] = df_top_10['Commissioning Date'].dt.year

# Create a pivot table to aggregate counts of occurrences of each combination of Manufacturer and Year
heatmap_data = df_top_10.groupby(['Manufacturer', 'Year']).size().unstack(fill_value=0)

# Create the heatmap with grid
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap='viridis', cbar_kws={'label': 'Frequency'}, linewidths=0.5)
plt.title('Heatmap of Top 10 Manufacturers and Commissioning Year')
plt.xlabel('Commissioning Year')
plt.ylabel('Manufacturer')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

### Summary

from this heat map, we can see that indeed Enercon has a very high frequency of wind parks being built consistently over the years. we see vestas indeed coming in second with a relatively high frequency over the years. from this graph, it looks like Siemens does very badly but given the context that they are the third largest producer of wind energy and that they have by far the highest number of large wind parks their frequency of production being down is quite understandable. These visualizations tell a very interesting story already.

```python
# Get the top 10 manufacturers by gross performance
top_10_manufacturers = df_gwt.groupby('Manufacturer')['Gross Performance'].sum().nlargest(10).index

# Filter the DataFrame to include only the top 10 manufacturers
df_top_10 = df_gwt[df_gwt['Manufacturer'].isin(top_10_manufacturers)]

# Extract the year from the 'Commissioning Date' column
df_top_10['Year'] = pd.to_datetime(df_top_10['Commissioning Date']).dt.year

# Group by 'Manufacturer' and 'Year', and calculate the sum of 'Gross Performance'
performance_by_manufacturer_year = df_top_10.groupby(['Manufacturer', 'Year'])['Gross Performance'].sum().reset_index()

# Create the point plot
plt.figure(figsize=(12, 8))
sns.pointplot(data=performance_by_manufacturer_year, x='Year', y='Gross Performance', hue='Manufacturer', markers='o')
plt.title('Sum of Gross Performance by Year for Top 10 Manufacturers')
plt.xlabel('Year')
plt.ylabel('Sum of Gross Performance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Manufacturer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### Summary

this graph represents the amount of production brought online in that year by that Manufacturer.

From this, we can see that Enerecon has consistently been bringing more production online consistently over the years. Vestas is a close second following a very similar trend. In 2015 we saw a large spike in production brought online by Siemens for 1 year it did better than any other manufacturer.

All things considered all these graphs agree with each other
